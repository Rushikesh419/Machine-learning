```python
circles
```

```python
sns.scatterplot(circles,x='X1',y='X2')
```

```python
def display(model,data):
  labels=model.fit_predict(data)
  sns.scatterplot(data=data,x='X1',y='X2',hue=labels,palette='tab10')
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=2)
```

```python
display(model,circles)
```

```python
from sklearn.cluster import DBSCAN
```

```python
model=DBSCAN()
```

```python
display(model,blobs)
```

```python
model=DBSCAN(eps=0.15)
display(model,circles)
```
