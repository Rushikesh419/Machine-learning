```python
pd.get_dummies(df.drop('species',axis=1),drop_first=True)
```

```python
X =pd.get_dummies(df.drop('species',axis=1),drop_first=True)
```

```python
y=df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model= DecisionTreeClassifier() # first by default
```

```python
model.fit(X_train,y_train)
```

```python
base_pred=model.predict(X_test)
```

```python
base_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report,confusion_matrix
```

```python
print(classification_report(y_test,base_pred))
```

```python
confusion_matrix(y_test,base_pred)
```

```python
ConfusionMatrixDisplay.from_estimator(model,X_test,y_test)
```

```python
model.feature_importances_
```

```python
X.columns
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_,
             columns=['feature_importance']).sort_values('feature_importance')
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(10,8),dpi=100)
plot_tree(model);
```

```python
# tree without column index but with column names
plt.figure(figsize=(10,8),dpi=100)
plot_tree(model,feature_names=X.columns,filled=True);
```

```python
# just a function for convinience

def report_model (model):
  model_preds=model.predict(X_test)
  print(classification_report(y_test,model_preds))
  print('\n')
  plt.figure(figsize=(10,8),dpi=100)
  plot_tree(model,feature_names=X.columns,filled=True);
```

```python
report_model(model)
```

**Exploring decision tree hyperparameters**

```python
pruned_tree=DecisionTreeClassifier(max_depth=2)
```

```python
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

```python
max_leaf_tree=DecisionTreeClassifier(max_leaf_nodes=3)
```

```python
max_leaf_tree.fit(X_train,y_train)
```

```python
report_model(max_leaf_tree)
```

```python
entropy_tree=DecisionTreeClassifier(criterion='entropy')
```

```python
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```

**just for the salke of it do a grid search model**

```python
param_grid={'criterion':['gini','entropy'],'max_depth':[2,3,4,5,6,7],'max_leaf_nodes':[2,3,4,5,6,7,8,9,10]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
dsc=DecisionTreeClassifier()
```

```python
grid=GridSearchCV(dsc,param_grid)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_pred=grid.predict(X_test)
```

```python
grid_pred
```

```python
confusion_matrix(y_test,grid_pred)
```

```python
ConfusionMatrixDisplay.from_estimator(grid,X_test,y_test)
```

```python
print(classification_report(y_test,grid_pred))
```

```python
# gridsearch classification treed

from sklearn.tree import export_graphviz
import graphviz
```

```python
dot_data = export_graphviz(grid.best_estimator_, out_file=None, filled=True, rounded=True,
                                class_names=['0','1','2'],feature_names=X.columns)
```
