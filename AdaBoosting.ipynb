### Mushroom Hunting: Edible or Poisonous?

Data Source: https://archive.ics.uci.edu/ml/datasets/Mushroom


This data set includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family (pp. 500-525). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like ``leaflets three, let it be'' for Poisonous Oak and Ivy.

# Goal

**THIS IS IMPORTANT, THIS IS NOT OUR TYPICAL PREDICTIVE MODEL!**

Our general goal here is to see if we can harness the power of machine learning and boosting to help create not just a predictive model, but a general guideline for features people should look out for when picking mushrooms.

```python
X=df.drop('class',axis=1)
```

```python
y=df['class']
```

```python
X=pd.get_dummies(X,drop_first=True)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
# simple
model=AdaBoostClassifier(n_estimators=1)
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay,accuracy_score
```

```python
pred=model.predict(X_test)
```

```python
pred
```

```python
print(classification_report(y_test,pred))
```

```python
confusion_matrix(y_test,pred)
```

```python
ConfusionMatrixDisplay.from_estimator(model,X_test,y_test)
```

```python
model.feature_importances_
```

```python
model.feature_importances_.argmax()
```

Analysis performance as we add more weak learners

```python
len(X.columns)
```

```python
error_rates=[]
for n in range(1,96):
  model=AdaBoostClassifier(n_estimators=n)
  model.fit(X_train,y_train)
  preds=model.predict(X_test)

  error=1-accuracy_score(y_test,preds)

  error_rates.append(error)
```

```python
plt.plot(range(1,96),error_rates)
```

```python
model
```

```python
model.feature_importances_
```

```python
features=pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Importance'])
```

```python
features.sort_values('Importance',ascending=False)
```

```python
imp_feats=features[features['Importance']>0]
```

```python
imp_feats
```

```python
plt.figure(figsize=(10,6),dpi=150)
sns.barplot(data=imp_feats.sort_values('Importance'),x=imp_feats.index,y='Importance',palette="pastel")
plt.xticks(rotation=90)
```
