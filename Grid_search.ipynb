```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



df1=pd.read_csv('/content/drive/MyDrive/Advertising.csv')


df=df1.drop('Unnamed: 0',axis=1)


X=df.drop('Sales',axis=1)


y=df['Sales']
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.30,random_state=101)


from sklearn.preprocessing import StandardScaler


scaler=StandardScaler()


scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)


X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elasic_net_model=ElasticNet()
```

```python
param_grid={'alpha':[0.1,1,5,10,50,100],
            'l1_ratio':[0.1,.5,.7,.95,.99,1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model=GridSearchCV(estimator=base_elasic_net_model,
                        param_grid=param_grid,
                        scoring='neg_mean_squared_error',cv=5,verbose=2) # verbos is the higer the more messages
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
# the actual result for other combination that was not best
grid_model.cv_results_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```
